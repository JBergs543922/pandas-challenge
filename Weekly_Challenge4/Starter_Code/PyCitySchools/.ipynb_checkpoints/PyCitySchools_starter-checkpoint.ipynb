# PyCity Schools Analysis

- school_summary Analysis:
- Our initial table does not provided much data. We can see that our mathematics test scores and pass rates lag behind reading scores and rates across the 15 schools in our school district.

- per_school_summary Analysis:
- Here we can examine some details, which could point to the discrepancy between math and reading scores/pass rates. 
- Firstly, the greater reading pass rate is consistent in the highest and lowest scoring schools. Math passing rates seem to lag behind reading across the board. 
- Second, we can observe that the total budgets and per capita budgets are not vastly different between each set of five schools. There are lower and high per capita values in both lists. However, a common factor among the highest performing schools is that they are charter, while the lowest performing schools are district.
- Third, the five highest schools seem to have a significantly lower population, than the five lowest performing schools. The lower passing rate might not be an issue of money and more one of student to staff ratio. This conclusion is supported by the groupings in the latter portion of the analysis. School size and type seem to make the biggest difference in passing scores.
---

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = len(school_data_complete['school_name'].unique()) #school_data_complete["school_name"].value_counts() counts the students at each school
school_count
```

```python
# Calculate the total number of students
student_count = float(len(school_data_complete))
student_count
```

```python
# Calculate the total budget
school_budget = school_data_complete[['school_name','budget']]
school_budgets = school_data_complete['budget'].unique()
school_budget_total = float(school_budgets.sum())
school_budget_total
#school_budget.head()
#total_budget = 
#total_budget = school_data_complete['budget'].sum()
#total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = round(school_data_complete['math_score'].mean(),2)
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = round(school_data_complete['reading_score'].mean(),2)
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = round(passing_math_count / float(student_count) * 100,2)
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = round(passing_reading_count / float(student_count)*100,2)
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = round(passing_math_reading_count /  float(student_count) * 100,2)
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame #this felt labor intensive, is there a better way?
#You can use groupby
district_summary = pd.DataFrame(
                                {'Total Schools':[school_count], 
                                 'Total Students':[student_count], 
                                 'Total Budget':[school_budget_total], 
                                 'Average Math Score':[average_math_score], 
                                 'Average Reading Score':[average_reading_score], 
                                 '% Passing Math':[passing_math_percentage], 
                                 '% Passing Reading':[passing_reading_percentage], 
                                 '% Overall Passing':[overall_passing_rate]})

# Formatting
district_summary["Total Students"] = district_summary['Total Students'].map("{:,}".format)
district_summary["Total Budget"] = district_summary['Total Budget'].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

## School Summary

```python
# Use the code provided to select all of the school types
school_types_total = school_data_complete[['school_name','type']].drop_duplicates().set_index('school_name')['type']
school_types_total
#school_types = school_types_total.value_counts()
#school_types
```

```python
# Calculate the total student count per school
per_school_counts = school_data_complete['school_name'].value_counts()
per_school_counts
```

```python
# Calculate the total school budget and per capita spending per school
per_school_budget_series = school_data_complete['budget'].unique()
per_school_budget_series = school_data_complete[['school_name','budget']].drop_duplicates().set_index('school_name')['budget']
per_school_budget_series
per_school_capita_df = per_school_budget_series/per_school_counts
per_school_capita_df
```

```python
# Calculate the average test scores per school for math and reading
grouped_df = school_data_complete.groupby('school_name')[['math_score','reading_score']]
per_school_math = round(grouped_df.mean()['math_score'],2)
per_school_reading = round(grouped_df.mean()['reading_score'],2)
#print(per_school_math)
#per_school_reading
```

```python
#combine what we have so far

summary = pd.concat([school_types_total,
                     per_school_counts,
                     per_school_budget_series,
                     per_school_capita_df,
                     per_school_math, 
                     per_school_reading], axis = 1)
summary.columns = ['School Type','Total Students',
                   'Total School Budget',
                   'Per Student Budget',
                   'Average Math Score', 
                   'Average Reading Score']

summary

#summary = pd.DataFrame([per_school_counts, per_school_math, per_school_reading]) doesn't cut it
# Formatting
#per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
#per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)
```

```python
# Calculate the number of students per school with reading scores of 70 or higher
students_passing_reading_overall = school_data_complete[school_data_complete['reading_score']>69]
students_passing_reading_perschool =  students_passing_reading_overall['school_name'].value_counts()
students_passing_reading_perschool.head()
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
school_students_passing_math_and_reading.head()
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = round(students_passing_math_perschool / per_school_counts * 100,2)
per_school_passing_reading = round(students_passing_reading_perschool / per_school_counts * 100,2)
overall_passing_rate = round(school_students_passing_math_and_reading / per_school_counts * 100,2)
#per_school_passing_math.head()
#per_school_passing_reading.head()
#overall_passing_rate.head()
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.

per_school_summary = pd.concat([school_types_total,per_school_counts,per_school_budget_series,per_school_capita_df,per_school_math, per_school_reading,per_school_passing_math,per_school_passing_reading, overall_passing_rate], axis = 1)
per_school_summary.columns = ['School Type','Total Students','Total School Budget','Per Student Budget',"Average Math Score", 'Average Reading Score','% Passing Math','%Passing Reading',"% Overall Passing"]
# Formatting
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

per_school_summary = per_school_summary.sort_values('school_name')

per_school_summary

#grouped_df = school_data_complete.groupby('school_name')['type']
#grouped_df.head()
#summary = pd.DataFrame([per_school_counts, per_school_math, per_school_reading]) doesn't cut it

#per_school_summary = pd.DataFrame(
                                #{'School Type':[school_types_total], 
                                 #'Total Students':[per_school_counts], 
                                 #'Total School Budget':[school_budget_total], 
                                 #'Per Student Budget':[per_school_capita_df],
                                 #'Average Math Score':[average_math_score], 
                                 #'Average Reading Score':[average_reading_score], 
                                 #'% Passing Math':[passing_math_percentage], 
                                 #'% Passing Reading':[passing_reading_percentage], 
                                 #'% Overall Passing':[overall_passing_rate]})

# Formatting
#per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
#per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)
#per_school_summary = pd.concat([school_types_total,
                                #per_school_counts,
                                #school_budget_total,
                                #per_school_capita_df,
                                #average_math_score,
                                #average_reading_score,
                                #passing_math_percentage,
                                #passing_reading_percentage,
                                #overall_passing_rate], axis = 1)
#per_school_summary.columns = ['School Type', 'Total Students','Total School Budget','Per Student Budget','Average Math Score','Average Reading Score','% Passing Math', '% Passing Reading','% Overall Passing']
#summary

# Display the DataFrame
#per_school_summary.head()
```

## Math Scores by Grade

```python
#an alternative way I used to group the grade means by school
grades_grouped = school_data_complete.groupby(['school_name','grade'])['math_score'].mean().to_frame()
grades_grouped
#grades_grouped.count()
```

```python
# Use the code provided to separate the data by grade groupby inside groupby
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

#ninth_graders
# Group by `school_name` and take the mean of the `math_score` column for each.
ninth_grader_math_scores = round(ninth_graders.groupby('school_name')[['math_score']].mean(),2)
#ninth_grade_math_scores_mean = ninth_grade_math_scores['math_score'].mean()
tenth_grader_math_scores = round(tenth_graders.groupby('school_name')[['math_score']].mean(),2)
eleventh_grader_math_scores = round(eleventh_graders.groupby('school_name')[['math_score']].mean(),2)
twelfth_grader_math_scores = round(twelfth_graders.groupby('school_name')[['math_score']].mean(),2)

# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.concat([ninth_grader_math_scores,tenth_grader_math_scores,eleventh_grader_math_scores,twelfth_grader_math_scores],axis = 1)
math_scores_by_grade.columns = ['9th','10th','11th','12th']

# Minor data wrangling
math_scores_by_grade.index.name = None

# Display the DataFrame
math_scores_by_grade
```

## Reading Score by Grade 

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group by `school_name` and take the mean of the the `reading_score` column for each.
ninth_grader_reading_scores = round(ninth_graders.groupby('school_name')[['reading_score']].mean(),2)
tenth_grader_reading_scores = round(tenth_graders.groupby('school_name')[['reading_score']].mean(),2)
eleventh_grader_reading_scores = round(eleventh_graders.groupby('school_name')[['reading_score']].mean(),2)
twelfth_grader_reading_scores = round(twelfth_graders.groupby('school_name')[['reading_score']].mean(),2)

# Combine each of the scores above into single DataFrame called `reading_scores_by_grade`
reading_scores_by_grade = pd.concat([ninth_grader_reading_scores,tenth_grader_reading_scores,eleventh_grader_reading_scores,twelfth_grader_reading_scores],axis = 1)
reading_scores_by_grade.columns = ['9th','10th','11th','12th']

# Minor data wrangling
reading_scores_by_grade = reading_scores_by_grade[["9th", "10th", "11th", "12th"]]
reading_scores_by_grade.index.name = None

# Display the DataFrame
reading_scores_by_grade
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
#school_spending_df = school_spending_df[['Total School Budget','Per Student Budget']].astype(float)
school_spending_df.head()
```

```python
# Use `pd.cut` to categorize spending based on the bins.

#had to use a dataframe that had not had the budget figures converted to strings
summary_per_stu_budget = summary[['Per Student Budget']]
#summary_per_stu_budget.columns = ["Per Student Budget"]
summary_per_stu_budget.head()
#school_spending_df["Spending Ranges (Per Student)"] = pd.cut[school_spending_df['Per Student Budget'],spending_bins,labels]
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df['Spending Ranges (Per Student)'] = pd.cut(summary['Per Student Budget'],bins = spending_bins, labels = labels)
#school_spend_df.drop('bins')
school_spending_df.sort_values('school_name')
school_spending_df
#school_spending_df["Spending Ranges (Per Student)"] = pd.cut('Per Student Budget',spending_bins,labels)
#summary_per_stu_budget['bins'].head()
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["%Passing Reading"].mean()
spending_overall_passing = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
```

```python
# Assemble into DataFrame
spending_summary = pd.concat([spending_math_scores,spending_reading_scores, spending_passing_math, spending_passing_reading,spending_overall_passing],axis=1)
spending_summary.columns = [['Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','% Overall Passing']]
# Display results
spending_summary
```

## Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

per_school_summary["School Size"] = pd.cut(x = summary['Total Students'],bins = size_bins, labels = labels)
per_school_summary
```

```python
# Calculate averages for the desired columns. 
size_math_scores = per_school_summary.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["% Passing Math"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["%Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["% Overall Passing"].mean()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.concat([size_math_scores,size_reading_scores, size_passing_math, size_passing_reading,size_overall_passing],axis=1)
spending_summary.columns = [['Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','% Overall Passing']]

# Display results
size_summary
```
